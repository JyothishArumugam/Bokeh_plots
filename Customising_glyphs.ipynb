##### Bokeh learning 


```python
import pandas as pd
from bokeh.plotting import ColumnDataSource,figure,show
from bokeh.io import output_notebook
df=pd.read_csv('iris.csv')
df.head()
```

* **make a **_scatter plot_ **with the above data frames**
* here we discuss about circular markers,
* for more markers [click here](http://bokeh.pydata.org/en/0.9.3/docs/reference/models/markers.html)

```python
#make a figure with names for x,y axis and title for the plot
plot=figure(x_axis_label='sepal_length',y_axis_label='Sepal_width',title='Scatter plot')
#make a circle as the markers for the points
plot.circle(df['SepalLengthCm'],df['PetalWidthCm'],color='red',size=5)
output_notebook()
show(plot)
```

```python
plot2=figure(x_axis_label='sepal',y_axis_label='petal',title='plot2')
source1=ColumnDataSource(df)
plot2.circle(x='SepalLengthCm',y='PetalLengthCm',source=source1,size=8,color=['red','blue'])
show(plot2)
```

##### selection and non selection plot


```python
plot3=figure(x_axis_label='sepal',y_axis_label='petal',title='Selection and non selection',tools='lasso_select')
plot3.circle(x='SepalLengthCm',y='PetalLengthCm',source=source1,selection_color='red',nonselection_color='green',nonselection_alpha=0.8)
# if non selection color is not provided it will be like blue.
show(plot3)
```

##### modelling using the hover tools


```python
from bokeh.models import HoverTool
plot4=figure(title='hover selection')
plot4.circle(x='SepalLengthCm',y='PetalLengthCm',source=source1,
            hover_fill_color='firebrick',hover_alpha=0.5,hover_line_color='green')
hover=HoverTool(tooltips='note',mode='vline')
plot4.add_tools(hover)
show(plot4)
```

```python
from bokeh.models import HoverTool
plot5=figure(title='hover selection')
plot5.circle(x='SepalLengthCm',y='PetalLengthCm',source=source1,
            hover_fill_color='firebrick',hover_alpha=0.5,hover_line_color='green')
hover=HoverTool(tooltips='note',mode='hline')
plot5.add_tools(hover)
show(plot5)
```

```python
from bokeh.models import HoverTool
plot6=figure(title='hover selection')
plot6.circle(x='SepalLengthCm',y='PetalLengthCm',source=source1,
            hover_fill_color='firebrick',hover_alpha=0.5,
             hover_line_color='green',
            line_color='red',
            fill_color='grey',alpha=0.1)
plot6.line(source=source1,x='SepalLengthCm',y='PetalLengthCm')
hover=HoverTool(tooltips=None,mode='hline')
plot6.add_tools(hover)
show(plot6)
```

```python
plot7=figure(title="colormapping")
from bokeh.models import CategoricalColorMapper
color_mapper=CategoricalColorMapper(factors=['Iris-setosa','Iris-versicolor','Iris-virginica']
                                   ,palette=['red','green','blue'])

plot7.circle(x='SepalLengthCm',y='PetalLengthCm',source=source1,
             color=dict(field='Species',transform=color_mapper),
            legend='Species')
show(plot7)
```
