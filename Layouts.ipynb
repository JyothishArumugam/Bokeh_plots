```python
import pandas as pd
df=pd.read_csv('iris.csv')
df.tail(5)
```

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure,show,ColumnDataSource
from bokeh.layouts import row
from bokeh.models import CategoricalColorMapper
source=ColumnDataSource(df)
color_mapper=CategoricalColorMapper(factors=['Iris-setosa','Iris-virginica','Iris-versicolor'],
                                   palette=['red','blue','green'])
```

```python
p1=figure(x_axis_label='Sepal Length',y_axis_label='Sepal Width',title='Sepal Length v/s Sepal Width',height=400,width=600)
p1.circle(x='SepalLengthCm',y='SepalWidthCm',size=5,source=source,
         color=dict(field='Species',transform=color_mapper),
         legend='Species')

p2=figure(x_axis_label='Sepal Length',y_axis_label='Petal Length',title='Sepal Length v/s Petal Length',height=400,width=600)
p2.circle(x='SepalLengthCm',y='PetalLengthCm',size=5,source=source,
          color=dict(field='Species',transform=color_mapper),
          legend='Species'
         )

p3=figure(x_axis_label='Sepal Length',y_axis_label='Petal Width',title='Sepal Length v/s Petal Width',height=400,width=600)
p3.circle(x='SepalLengthCm',y='PetalWidthCm',size=5,source=source,
         color=dict(field='Species',transform=color_mapper),
         legend='Species')


layout=row(p1,p2,p3)
output_notebook()
show(layout)
```

###### displaying in columns


```python
from bokeh.layouts import column
col_layout=column(p1,p2,p3)
show(col_layout)
```

#### nesting of things


```python
# Make a column layout that will be used as the second row: row2
row2 = column([p1, p2], sizing_mode='scale_width')

# Make a row layout that includes the above column layout: layout
layout22 = row([p3, row2], sizing_mode='scale_width')
show(layout22)
```

```python
# Make a column layout that will be used as the second row: row2
row2 = column([p1, p2], sizing_mode='scale_width')

# Make a row layout that includes the above column layout: layout
layout22 = row([row2,p3], sizing_mode='scale_width')
show(layout22)
```

#### Advanced Layouts
__Grid layout__&&&&
__Tabbed layout__

```python
from bokeh.layouts import gridplot
layoutg=gridplot([[None,p2],[p1,p3]],toolbar_location='left')
# argument accepts a list of lists, each list is a row in the grid , the toolbar location can be ="left,right, bottom, above,None"
show(layoutg)
```

#### Lets check the tabbed plots

```python
from bokeh.models.widgets import Panel,Tabs
#creating a panel for each tab
first_panel=Panel(title='first',child=row(p1,p2)) 
second_panel=Panel(title='second',child=row(p3))
#put te panels into the tabs
layoutt=Tabs(tabs=[first_panel,second_panel])
show(layoutt)
```
